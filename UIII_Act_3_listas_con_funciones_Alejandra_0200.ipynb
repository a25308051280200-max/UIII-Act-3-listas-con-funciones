### Ejemplo listas con funciones

Las listas con funciones en Python permiten organizar y manipular información de manera más sencilla y ordenada. En el código del sistema de deportistas, la lista llamada deportistas almacena varios registros usando diccionarios con datos como nombre, deporte, edad y país.

Cada función tiene una tarea específica: agregarDeportista() guarda nuevos datos, listarDeportistas() muestra todos los registros, buscarDeportista() encuentra un deportista por su nombre, modificarDeportista() actualiza la información y borrarDeportista() elimina registros. Esto ayuda a que el programa sea más organizado, reutilizable y fácil de entender.

El uso de listas y funciones es muy importante en programación porque permite manejar grandes cantidades de información de forma eficiente y facilita la creación de sistemas más completos y dinámicos.

```python
# ============================================
# SISTEMA DE GESTIÓN DE DEPORTISTAS
# ============================================

# Lista principal
deportistas = []


# ============================================
# FUNCIÓN: CARGAR 4 DEPORTISTAS
# ============================================
def leerDeportistas():

    print("\n=== INGRESO DE 4 DEPORTISTAS ===")

    for i in range(4):

        print(f"\nDeportista #{i+1}")

        nombre = input("Nombre: ")
        deporte = input("Deporte: ")
        edad = int(input("Edad: "))
        pais = input("País: ")

        deportista = {
            "nombre": nombre,
            "deporte": deporte,
            "edad": edad,
            "pais": pais
        }

        deportistas.append(deportista)

    print("\n✅ Los 4 deportistas fueron registrados.")


# ============================================
# FUNCIÓN: AGREGAR
# ============================================
def agregarDeportista():

    print("\n--- AGREGAR DEPORTISTA ---")

    nombre = input("Nombre: ")
    deporte = input("Deporte: ")
    edad = int(input("Edad: "))
    pais = input("País: ")

    deportista = {
        "nombre": nombre,
        "deporte": deporte,
        "edad": edad,
        "pais": pais
    }

    deportistas.append(deportista)

    print("✅ Deportista agregado correctamente.")


# ============================================
# FUNCIÓN: LISTAR
# ============================================
def listarDeportistas():

    print("\n--- LISTA DE DEPORTISTAS ---")

    if len(deportistas) == 0:
        print("No hay deportistas registrados.")

    else:
        for i, dep in enumerate(deportistas):

            print(f"\nRegistro #{i+1}")
            print("Nombre :", dep["nombre"])
            print("Deporte:", dep["deporte"])
            print("Edad   :", dep["edad"])
            print("País   :", dep["pais"])


# ============================================
# FUNCIÓN: BUSCAR
# ============================================
def buscarDeportista():

    print("\n--- BUSCAR DEPORTISTA ---")

    nombre_buscar = input("Ingrese el nombre: ")

    encontrado = False

    for dep in deportistas:

        if dep["nombre"].lower() == nombre_buscar.lower():

            print("\n✅ Deportista encontrado")
            print("Nombre :", dep["nombre"])
            print("Deporte:", dep["deporte"])
            print("Edad   :", dep["edad"])
            print("País   :", dep["pais"])

            encontrado = True
            break

    if encontrado == False:
        print("❌ Deportista no encontrado.")


# ============================================
# FUNCIÓN: MODIFICAR
# ============================================
def modificarDeportista():

    print("\n--- MODIFICAR DEPORTISTA ---")

    nombre_buscar = input("Ingrese el nombre: ")

    for dep in deportistas:

        if dep["nombre"].lower() == nombre_buscar.lower():

            print("\nIngrese los nuevos datos")

            dep["nombre"] = input("Nuevo nombre: ")
            dep["deporte"] = input("Nuevo deporte: ")
            dep["edad"] = int(input("Nueva edad: "))
            dep["pais"] = input("Nuevo país: ")

            print("✅ Deportista modificado.")
            return

    print("❌ Deportista no encontrado.")


# ============================================
# FUNCIÓN: BORRAR
# ============================================
def borrarDeportista():

    print("\n--- BORRAR DEPORTISTA ---")

    nombre_buscar = input("Ingrese el nombre: ")

    for dep in deportistas:

        if dep["nombre"].lower() == nombre_buscar.lower():

            deportistas.remove(dep)

            print("✅ Deportista eliminado.")
            return

    print("❌ Deportista no encontrado.")


# ============================================
# FUNCIÓN: MENÚ
# ============================================
def menu():

    print("\n================================")
    print("   SISTEMA DE DEPORTISTAS")
    print("================================")
    print("1. Agregar deportista")
    print("2. Listar deportistas")
    print("3. Buscar deportista")
    print("4. Modificar deportista")
    print("5. Borrar deportista")
    print("6. Salir")
    print("================================")


# ============================================
# PRIMERO SE LEEN LOS 4 DEPORTISTAS
# ============================================
leerDeportistas()


# ============================================
# PROGRAMA PRINCIPAL
# ============================================
while True:

    menu()

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregarDeportista()

    elif opcion == "2":
        listarDeportistas()

    elif opcion == "3":
        buscarDeportista()

    elif opcion == "4":
        modificarDeportista()

    elif opcion == "5":
        borrarDeportista()

    elif opcion == "6":
        print("👋 Saliendo del sistema...")
        break

    else:
        print("❌ Opción inválida")
```
