### Ejemplo listas con funciones

Las listas con funciones en Python permiten organizar y manipular información de manera más sencilla y ordenada. En el código del sistema de deportistas, la lista llamada deportistas almacena varios registros usando diccionarios con datos como nombre, deporte, edad y país.

Cada función tiene una tarea específica: agregarDeportista() guarda nuevos datos, listarDeportistas() muestra todos los registros, buscarDeportista() encuentra un deportista por su nombre, modificarDeportista() actualiza la información y borrarDeportista() elimina registros. Esto ayuda a que el programa sea más organizado, reutilizable y fácil de entender.

El uso de listas y funciones es muy importante en programación porque permite manejar grandes cantidades de información de forma eficiente y facilita la creación de sistemas más completos y dinámicos.

```python
# ==========================================
# CBTIS 128
# Submódulo 2 - Elabora proyectos de programación lógica
# Sistema de lista de deportistas usando funciones
# ==========================================

# Lista principal
deportistas = []

# ------------------------------------------
# Función para agregar deportistas
# ------------------------------------------
def agregarDeportista():
    nombre = input("Nombre del deportista: ")
    deporte = input("Deporte: ")
    edad = int(input("Edad: "))

    deportista = {
        "nombre": nombre,
        "deporte": deporte,
        "edad": edad
    }

    deportistas.append(deportista)
    print("✅ Deportista agregado correctamente\n")


# ------------------------------------------
# Función para leer datos
# ------------------------------------------
def leerDeportista():
    nombre = input("Ingresa el nombre del deportista: ")

    for d in deportistas:
        if d["nombre"].lower() == nombre.lower():
            print("\n--- DATOS DEL DEPORTISTA ---")
            print("Nombre:", d["nombre"])
            print("Deporte:", d["deporte"])
            print("Edad:", d["edad"])
            return

    print("❌ Deportista no encontrado\n")


# ------------------------------------------
# Función para listar deportistas
# ------------------------------------------
def listarDeportistas():

    if len(deportistas) == 0:
        print("No hay deportistas registrados\n")
    else:
        print("\n===== LISTA DE DEPORTISTAS =====")

        for i, d in enumerate(deportistas, start=1):
            print(f"{i}. {d['nombre']} - {d['deporte']} - {d['edad']} años")

        print()


# ------------------------------------------
# Función para buscar deportista
# ------------------------------------------
def buscarDeportista():
    nombre = input("Nombre a buscar: ")

    encontrado = False

    for d in deportistas:
        if d["nombre"].lower() == nombre.lower():
            print("\n✅ Deportista encontrado")
            print("Nombre:", d["nombre"])
            print("Deporte:", d["deporte"])
            print("Edad:", d["edad"])
            encontrado = True

    if not encontrado:
        print("❌ Deportista no encontrado\n")


# ------------------------------------------
# Función para modificar deportista
# ------------------------------------------
def modificarDeportista():
    nombre = input("Nombre del deportista a modificar: ")

    for d in deportistas:
        if d["nombre"].lower() == nombre.lower():

            print("\nIngrese los nuevos datos")

            d["nombre"] = input("Nuevo nombre: ")
            d["deporte"] = input("Nuevo deporte: ")
            d["edad"] = int(input("Nueva edad: "))

            print("✅ Datos modificados correctamente\n")
            return

    print("❌ Deportista no encontrado\n")


# ------------------------------------------
# Función para borrar deportista
# ------------------------------------------
def borrarDeportista():
    nombre = input("Nombre del deportista a borrar: ")

    for d in deportistas:
        if d["nombre"].lower() == nombre.lower():
            deportistas.remove(d)
            print("✅ Deportista eliminado correctamente\n")
            return

    print("❌ Deportista no encontrado\n")


# ------------------------------------------
# MENÚ PRINCIPAL
# ------------------------------------------
def menuLista():

    opcion = 0

    while opcion != 7:

        print("====== MENÚ LISTA DE DEPORTISTAS ======")
        print("1. Agregar deportista")
        print("2. Leer deportista")
        print("3. Listar deportistas")
        print("4. Buscar deportista")
        print("5. Modificar deportista")
        print("6. Borrar deportista")
        print("7. Salir")

        opcion = int(input("Selecciona una opción: "))

        if opcion == 1:
            agregarDeportista()

        elif opcion == 2:
            leerDeportista()

        elif opcion == 3:
            listarDeportistas()

        elif opcion == 4:
            buscarDeportista()

        elif opcion == 5:
            modificarDeportista()

        elif opcion == 6:
            borrarDeportista()

        elif opcion == 7:
            print("Programa finalizado")

        else:
            print("❌ Opción incorrecta\n")


# Ejecutar programa
menuLista()
print("Elaborado por Alejandra Rivera NL=45")
```
