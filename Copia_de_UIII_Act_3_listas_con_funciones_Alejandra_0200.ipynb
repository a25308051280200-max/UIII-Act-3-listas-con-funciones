### Ejemplo listas con funciones

Las listas con funciones en Python permiten organizar y manipular información de manera más sencilla y ordenada. En el código del sistema de deportistas, la lista llamada deportistas almacena varios registros usando diccionarios con datos como nombre, deporte, edad y país.

Cada función tiene una tarea específica: agregarDeportista() guarda nuevos datos, listarDeportistas() muestra todos los registros, buscarDeportista() encuentra un deportista por su nombre, modificarDeportista() actualiza la información y borrarDeportista() elimina registros. Esto ayuda a que el programa sea más organizado, reutilizable y fácil de entender.

El uso de listas y funciones es muy importante en programación porque permite manejar grandes cantidades de información de forma eficiente y facilita la creación de sistemas más completos y dinámicos.

```python
# ==========================================
# LISTA DE DEPORTISTAS
# Uso de funciones y listas
# ==========================================

# Lista principal
deportistas = []

# ------------------------------------------
# Función agregar
# ------------------------------------------
def agregarDeportista():
    nombre = input("Ingresa el nombre del deportista: ")
    deportistas.append(nombre)
    print("✅ Deportista agregado correctamente\n")


# ------------------------------------------
# Función leer
# ------------------------------------------
def leerDeportista():

    if len(deportistas) == 0:
        print("❌ No hay deportistas registrados\n")
    else:
        posicion = int(input("Ingresa la posición del deportista: "))

        if posicion >= 0 and posicion < len(deportistas):
            print("Deportista:", deportistas[posicion], "\n")
        else:
            print("❌ Posición incorrecta\n")


# ------------------------------------------
# Función listar
# ------------------------------------------
def listarDeportistas():

    if len(deportistas) == 0:
        print("❌ No hay deportistas registrados\n")
    else:
        print("\n===== LISTA DE DEPORTISTAS =====")

        for i in range(len(deportistas)):
            print(i, "-", deportistas[i])

        print()


# ------------------------------------------
# Función buscar
# ------------------------------------------
def buscarDeportista():
    nombre = input("Ingresa el nombre a buscar: ")

    if nombre in deportistas:
        print("✅ Deportista encontrado\n")
    else:
        print("❌ Deportista no encontrado\n")


# ------------------------------------------
# Función modificar
# ------------------------------------------
def modificarDeportista():

    nombre = input("Ingresa el nombre a modificar: ")

    if nombre in deportistas:
        posicion = deportistas.index(nombre)

        nuevo = input("Ingresa el nuevo nombre: ")

        deportistas[posicion] = nuevo

        print("✅ Deportista modificado correctamente\n")

    else:
        print("❌ Deportista no encontrado\n")


# ------------------------------------------
# Función borrar
# ------------------------------------------
def borrarDeportista():

    nombre = input("Ingresa el nombre a borrar: ")

    if nombre in deportistas:
        deportistas.remove(nombre)
        print("✅ Deportista eliminado correctamente\n")
    else:
        print("❌ Deportista no encontrado\n")


# ------------------------------------------
# MENÚ
# ------------------------------------------
def menuLista():

    opcion = 0

    while opcion != 7:

        print("====== MENÚ DEPORTISTAS ======")
        print("1. Agregar deportista")
        print("2. Leer deportista")
        print("3. Listar deportistas")
        print("4. Buscar deportista")
        print("5. Modificar deportista")
        print("6. Borrar deportista")
        print("7. Salir")

        opcion = int(input("Selecciona una opción: "))

        if opcion == 1:
            agregarDeportista()

        elif opcion == 2:
            leerDeportista()

        elif opcion == 3:
            listarDeportistas()

        elif opcion == 4:
            buscarDeportista()

        elif opcion == 5:
            modificarDeportista()

        elif opcion == 6:
            borrarDeportista()

        elif opcion == 7:
            print("Programa finalizado")

        else:
            print("❌ Opción incorrecta\n")


# Ejecutar menú
menuLista()
```
